## Run Reporting Tests

Execute the focused pytest file and capture the output.

```python
import sys
import pathlib

print('sys.executable:', sys.executable)
print('sys.path:')
for p in sys.path:
    print(' ', p)

try:
    import pandas as pd
    print('pandas module file:', getattr(pd, '__file__', None))
    print('pandas has DataFrame:', hasattr(pd, 'DataFrame'))
    print('pandas dir sample:', [name for name in dir(pd) if 'DataFrame' in name or name.startswith('__')][:20])
except Exception as exc:
    print('pandas import problem:', type(exc).__name__, exc)
```

```python
import os
from pathlib import Path

print('site-packages paths:')
for p in sys.path:
    if 'site-packages' in str(p).lower():
        print(' -', p)
        p = Path(p)
        if p.exists():
            for child in sorted(p.glob('pandas*')):
                print('   pandas entry:', child, child.is_dir(), child.is_file())

try:
    import pandas as pd
    print('pandas __path__:', getattr(pd, '__path__', None))
    if getattr(pd, '__path__', None) is not None:
        for p in pd.__path__:
            print('  path item:', p)
            print('  contents:', list(Path(p).glob('*'))[:20])
except Exception as exc:
    print('import pandas failed:', repr(exc))
```
